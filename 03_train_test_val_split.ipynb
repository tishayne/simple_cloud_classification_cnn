```python
import os
import shutil
from sklearn.model_selection import train_test_split

''' Manually label clear, cloudy and overcast images and save them in e.g. data/dataset_split/clear''' 
DATASET_DIR = "data/dataset_split" # path to folder containing clear, cloudy and overcast folder.
OUTPUT_DIR = "data/dataset" # This is where the splitted dataset will be.

classes = ["clear", "cloudy", "overcast"]

splits = ["train", "val", "test"]

# Create 3 folders
for split in splits:
    for cls in classes:
        os.makedirs(os.path.join(OUTPUT_DIR, split, cls), exist_ok=True)

for cls in classes:
    class_dir = os.path.join(DATASET_DIR, cls)
    images = [f for f in os.listdir(class_dir) if f.lower().endswith((".jpg", ".jpeg", ".png"))]

    train_imgs, temp_imgs = train_test_split(images, test_size=0.2, shuffle=True, random_state=42)

    val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.5, shuffle=True, random_state=42)

    for img in train_imgs:
        shutil.copy(os.path.join(class_dir, img), os.path.join(OUTPUT_DIR, "train", cls, img))

    for img in val_imgs:
        shutil.copy(os.path.join(class_dir, img), os.path.join(OUTPUT_DIR, "val", cls, img))

    for img in test_imgs:
        shutil.copy(os.path.join(class_dir, img), os.path.join(OUTPUT_DIR, "test", cls, img))

    print(f"{cls}: {len(train_imgs)} train, {len(val_imgs)} val, {len(test_imgs)} test")
```
